```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("genz_social_media_usage_1M.csv")
```

```python
print("Preview:\n",df.head())
print("\n Total Users: \n", len(df))
```

```python
df.columns = df.columns.str.strip().str.lower()
df.dropna(inplace=True)
df.rename(columns={'primary_platform':'platform'},inplace=True)

df
```

```python
addiction_map={'Low':1,'Medium':2,'High':3}
df['addiction_score']=df['addiction_level'].map(addiction_map)

df['heavy_user']=np.where(df['daily_usage_hours']> 5, 1, 0)
df

```

```python
def age_category(age):
    if age>=10 and age<=15:
        return "Teen"
    elif age>=16 and age<=18:
        return "Late Teen"
    elif age>=19 and age<=22:
        return "Young Adult"
    elif age>=23 and age<=30:
        return "Adult"
def addiction_risk(usage):
    if usage<3:
        return "Low"
    elif usage>=3 and usage<=6:
        return "Medium"
    elif usage>6:
        return "High"
    
```

```python
df['age_category']= df['age'].apply(age_category)
df['addiction_risk']=df['daily_usage_hours'].apply(addiction_risk)
df
```

```python
df.drop_duplicates()
df
```

```python
avg_daily_hours = df['daily_usage_hours'].mean()
print(avg_daily_hours)
```

```python
def get_most_used_platform(platform_col):
    counts = platform_col.value_counts()
    print(counts) 
    return counts.index[0] 

most_used = get_most_used_platform(df['platform'])
print(most_used)
```

```python
platform_counts = df['platform'].value_counts()

most_used_platform = platform_counts.idxmax()   # platform name
max_count = platform_counts.max()               # maximum count

print("Platform:", most_used_platform)
print("Count:", max_count)
```
